```python
import pandas as pd
```

```python
data = pd.read_csv('decomposition_results.csv')
```

```python
data.head()
```

```python
d_subject = data.copy()
d_subject['is_pinch'] = 0
d_subject['is_grip'] = 0
d_subject.loc[d_subject['Gesture'] == 'pinch', 'is_pinch'] = 1
d_subject.loc[d_subject['Gesture'] == 'grip', 'is_grip'] = 1
agg_funs = ['mean', 'std']
d_subject_grouped = d_subject.groupby('Subject').agg({'CoV': agg_funs, 'MDR': agg_funs,
                                                      'Pulses': agg_funs, 'is_pinch': 'sum', 'is_grip': 'sum'}).round(2)

d_subject_grouped
```

```python
d_subject_grouped['CoV_tmp'] = '$' + d_subject_grouped['CoV']['mean'].astype(str) + '\pm' + d_subject_grouped['CoV']['std'].astype(str) + '$'
d_subject_grouped['MDR_tmp'] = '$' + d_subject_grouped['MDR']['mean'].astype(str) + '\pm' + d_subject_grouped['MDR']['std'].astype(str) + '$'
d_subject_grouped['Pulses_tmp'] = '$' + d_subject_grouped['Pulses']['mean'].astype(str) + '\pm' + d_subject_grouped['Pulses']['std'].astype(str) + '$'
d_subject_grouped['MUs (pinch/grip)'] = (d_subject_grouped['is_pinch']['sum'] + d_subject_grouped['is_grip']['sum']).astype(str) + ' (' + d_subject_grouped['is_pinch']['sum'].astype(str) + '/' + d_subject_grouped['is_grip']['sum'].astype(str) + ')'

col_names = ['CoV', 'MDR', 'Pulses']
d_subject_grouped.drop(columns=col_names + ['is_pinch', 'is_grip'], inplace=True)

d_subject_grouped.columns = col_names + ['MUs (pinch/grip)']
d_subject_grouped.reset_index(inplace=True)

d_subject_grouped
```

```python
print(d_subject_grouped[['Subject', 'MUs (pinch/grip)', 'Pulses', 'CoV', 'MDR']].to_latex(index=False, escape=False))
```

```python
d_hand = data.copy()
d_hand.drop(columns=['Subject'], inplace=True)
d_hand['MUs'] = 1

d_hand_grouped = d_hand.groupby('Gesture').agg({'CoV': agg_funs, 'MDR': agg_funs,
                                                'Pulses': agg_funs, 'MUs': 'sum'}).round(2)
d_hand_grouped
```

```python
d_hand_grouped['CoV_tmp'] = '$' + d_hand_grouped['CoV']['mean'].astype(str) + '\pm' + d_hand_grouped['CoV']['std'].astype(str) + '$'
d_hand_grouped['MDR_tmp'] = '$' + d_hand_grouped['MDR']['mean'].astype(str) + '\pm' + d_hand_grouped['MDR']['std'].astype(str) + '$'
d_hand_grouped['Pulses_tmp'] = '$' + d_hand_grouped['Pulses']['mean'].astype(str) + '\pm' + d_hand_grouped['Pulses']['std'].astype(str) + '$'
d_hand_grouped['MUs_tmp'] = d_hand_grouped['MUs']['sum']

col_names = ['CoV', 'MDR', 'Pulses', 'MUs']
d_hand_grouped.drop(columns=col_names, inplace=True)

d_hand_grouped.columns = col_names
d_hand_grouped.reset_index(inplace=True)

d_hand_grouped
```

```python
print(d_hand_grouped[['Gesture', 'MUs', 'Pulses', 'CoV', 'MDR']].to_latex(index=False, escape=False))
```
