```python
import pandas as pd
```

```python
data = pd.read_csv('tracking_results.csv')
data.head()
```

```python
d_subject = data.copy()
d_subject.drop(columns=['Pulses'], inplace=True)

d_subject['is_mixed'] = d_subject['is_grip'] = d_subject['is_pinch'] = 0
d_subject['sum'] = 1

d_subject.loc[d_subject['Source Type'] == 'mixed', 'is_mixed'] = 1
d_subject.loc[d_subject['Source Type'] == 'grip', 'is_grip'] = 1
d_subject.loc[d_subject['Source Type'] == 'pinch', 'is_pinch'] = 1

d_subject.head(10)
```

```python
d_subject_grouped = d_subject.groupby('Subject').agg('sum')
d_subject_grouped.reset_index()
d_subject_grouped.columns = ['Mixed', 'Grip', 'Pinch', 'Total']


d_subject_grouped = pd.concat({'MUs': d_subject_grouped}, axis=1)


d_subject_grouped
```

```python
print(d_subject_grouped.to_latex())
```

```python
d_type = data.copy()
d_type.drop(columns=['Subject'], inplace=True)
d_type['MUs'] = 1

d_type_grouped = d_type.groupby('Source Type').agg({'Pulses': ['mean', 'std'], 'MUs': 'sum'}).round(2)

d_type_grouped['Pulses_tmp'] = '$' + d_type_grouped['Pulses']['mean'].astype(str) + '\pm' + d_type_grouped['Pulses']['std'].astype(str) + '$'
d_type_grouped['MUs_tmp'] = d_type_grouped['MUs']['sum']

col_names = ['Pulses', 'MUs']
d_type_grouped.drop(columns=col_names, inplace=True)
d_type_grouped.columns = col_names

d_type_grouped.reset_index(inplace=True)

d_type_grouped
```

```python
print(d_type_grouped[['Source Type', 'MUs', 'Pulses']].to_latex(index=False, escape=False))
```
